```python
import pandas as pd
import duckdb
```

frida_amazon_products.parquet is a file which contains information about the products scraped from Amazon

```python
df = pd.read_parquet('frida_amazon_products.parquet')
df
```

### Doing a select query on parquet file

```python
duckdb.query(f'''
SELECT * FROM 'frida_amazon_products.parquet' where rating < 4 LIMIT 5
''').df()
```

SELECT query on parquet file was succesful

### Now doing DELETE/UPDATE query on the parquet file

```python
duckdb.query(f'''
UPDATE 'frida_amazon_products.parquet' SET rating = 5 WHERE platfrom=='Amazon'
''').df()
```
